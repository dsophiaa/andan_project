```python
import pandas as pd
import numpy as np
```

Самое первое что происходит - мы скачиваем рандомный документ оттуда и узнаём сколько и какие регионы + за какие года данные встречаются в таких документах 
Т.к. это единичный случай,  то мы там практически построчно пишем откуда и докуда идут регионы
После чего формируем словарь "тема": "ссылка на документ" и создаём начальный пустой дф где индексы это пары регион-год
Далее берём каждую ссылку, грубо обрезаем, чтобы получить примерный датафрейм данных, после чего проходясь по всем парам регион-год пытаемся найти по этим индексами значения в скачанных датасетах
В последнем файле у нас данные записаны по 3 категориям (все население, муж, жен) поэтому обрабатываем этот случай условием если в полученном датасете строк больше чем кол-во регионов
Далее обработка NaN-ов. Если у нас столбец, в котором данные отсутствуют для всего региона, то просто заполняем эти наны средним значением по столбцу
Если же у нас отсутствует какой-то конкретный код у региона, то считаем среднее по этому региону и заменяем 


```python
index_to_delete = ['Центральный федеральный округ',
                   'Северо-Западный федеральный округ',
                   "Южный \nфедеральный округ",
                   "Северо-Кавказский\nфедеральный округ",
                   "Приволжский федеральный округ",
                   "Уральский федеральный округ",
                  "Сибирский федеральный округ",
                  "Дальневосточный федеральный округ"]

df = pd.read_excel('https://rosstat.gov.ru/storage//2023/04-19/3ChO60pb/RR_pokaz_06-01_2022.xlsx')
years = df.iloc[8][7:-1].apply(int).values
df = df.dropna(axis=0, how='any')
df = df[~df['Unnamed: 0'].isin(index_to_delete)]
regions = df['Unnamed: 0'].values
```

```python
regions
```

```python
data = {
    'Число больничных коек': "https://rosstat.gov.ru/storage//2023/04-19/3ChO60pb/RR_pokaz_06-01_2022.xlsx",
    'Мощность амбулаторно-поликлинических организаций': 'https://rosstat.gov.ru/storage//2023/04-19/3ChO60pb/RR_pokaz_06-03_2022.xlsx',
    'Численность врачей всех специальностей': 'https://rosstat.gov.ru/storage//2023/04-19/3ChO60pb/RR_pokaz_06-04_2022.xlsx',
    'Численность среднего медицинского персонала': 'https://rosstat.gov.ru/storage//2023/04-19/3ChO60pb/RR_pokaz_06-06_2022.xlsx',
    'Заболеваемость на 1000 человек населения': 'https://rosstat.gov.ru/storage//2023/04-19/3ChO60pb/RR_pokaz_06-08_2022.xlsx',
    'Среднедушевые денежные доходы населения': 'https://rosstat.gov.ru/storage//2023/04-19/zWBpBEse/RR_pokaz_04-04_2022.xlsx',
    'Смертность населения в трудоспособном возрасте': 'https://rosstat.gov.ru/storage//2023/04-19/Tdt5j4g3/RR_pokaz_02-10_2022.xlsx',
    'Число предприятий и организаций': 'https://rosstat.gov.ru/storage//2023/04-19/gVL3NtVW/RR_pokaz_12-01_2022.xlsx',
    'Выбросы загрязняющих веществ в атмосферный воздух': 'http://rosstat.gov.ru/storage//2023/04-19/dvtmDkVj/RR_pokaz_08-03_2022.xlsx',
    'Улавливание загрязняющих атмосферу веществ': 'https://rosstat.gov.ru/storage//2023/04-19/dvtmDkVj/RR_pokaz_08-04_2022.xlsx',
    'Сброс загрязненных сточных вод в поверхностные водные объекты': 'https://rosstat.gov.ru/storage//2023/04-19/dvtmDkVj/RR_pokaz_08-08_2022.xlsx' ,
    'Средняя продолжительность жизни' : 'http://rosstat.gov.ru/storage//2023/04-19/Tdt5j4g3/RR_pokaz_02-15_2022.xlsx'
}
```

```python
all_dataset = pd.DataFrame()
all_dataset.index = regions
all_dataset['Year'] = str(list(years))
all_dataset['Year'] = all_dataset['Year'].apply(eval)
all_dataset = all_dataset.explode('Year')
all_dataset = all_dataset.reset_index().set_index(['index', 'Year'])
all_dataset
```

```python
list(years)
```

```python
for k, v in data.items():
    df = pd.read_excel(v)
    df = df[df.isin(regions).any(axis=1)]
    if len(df.columns) == 13:
        df.columns = ['Регион'] + list(years * 2) + list(years)
    elif len(df.columns) == 14:
        df.columns = ['Регион'] + list(years * 2) + list(years) + ['Место в РФ']
    elif len(df.columns) == 7:
        df.columns = ['Регион'] + list(years)
    else:
        df.columns = ['Регион'] + list(years) + ['Место в РФ']
    new_column = []
    df = df.set_index('Регион')
    for reg, year in all_dataset.index:
        try:
            if len(df) > len(regions):
                new_column.append(float(str(df.loc[reg][year][0]).replace(',', '.').replace('\xa0', '')))
            else:
                new_column.append(float(str(df.loc[reg][year]).replace(',', '.').replace('\xa0', '')))
        except Exception as e:
            new_column.append(np.nan)
    all_dataset[k] = new_column
```

```python
all_dataset = all_dataset.reset_index().rename(columns={'index': 'Регион', 'Year': 'Год'})
```

```python
has_na = all_dataset.groupby('Регион').apply(lambda x: x.isna().sum(axis=0))
has_na
```

```python
indexs, cols = (has_na == 6).values.nonzero()
indexs = has_na.index[indexs]
cols = has_na.columns[cols]
cols
```

```python
mean_vals = all_dataset.drop('Регион', axis=1).mean(axis=0)
median_vals = all_dataset.drop('Регион', axis=1).median(axis=0)

for i in range(len(cols)):
    if cols[i] in ['Число предприятий и организаций']:
        all_dataset.loc[all_dataset['Регион'] == indexs[i], cols[i]] = median_vals[cols[i]]
    else:
        all_dataset.loc[all_dataset['Регион'] == indexs[i], cols[i]] = mean_vals[cols[i]]
    
```
